### <span style="color: #1AA3D9"><center> Item 1: Diferenciación

Encuentre una expresión para la magnitud de la velocidad de la partícula en función del tiempo. Recuerde que la velocidad es la derivada de la posición, respecto al tiempo.

```python
!pip install sympy
!pip install numpy 
!pip install matplotlib
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math

t, a = sp.symbols("t, a")
x = a**(t/math.pi) * sp.sin(t)
y = a**(t/math.pi) * sp.cos(t)

velocidad_x = sp.diff(x, t)
velocidad_y = sp.diff(y, t)

v = sp.sqrt(sp.diff(x, t)**2 + sp.diff(y, t)**2)
print("Velocidad =", v)

a_val = float(input("Introduzca el valor de la constante a: "))
f_v = lambda tt: np.sqrt(
    (a_val**(tt/math.pi) * np.log(a_val)/math.pi * np.sin(tt) + a_val**(tt/np.pi) * np.cos(tt))**2 +
    (a_val**(tt/math.pi) * np.log(a_val)/math.pi * np.cos(tt) - a_val**(tt/np.pi) * np.sin(tt))**2)

tiempo = np.linspace(0, 10, 100)
vel = f_v(tiempo)

plt.plot(tiempo, vel)
plt.title("Magnitud de la velocidad")
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
plt.grid(True)
plt.show()
```

### <span style="color: #1AA3D9"><center> Item 2: Integración

Usando integración simbólica, encuentre una expresión para la distancia recorrida por la partícula en función del tiempo.

```python
pip install sympy
```

```python
import sympy as sp


t, tau, a = sp.symbols('t tau a', positive=True)


x = a**(tau/sp.pi) * sp.sin(tau)



y = a**(tau/sp.pi) * sp.cos(tau)


dx = sp.diff(x, tau)



dy = sp.diff(y, tau)


v = sp.sqrt(dx**2 + dy**2)


s = sp.integrate(v, (tau, 0, t))


print("Velocidad simplificada:")

sp.pprint(sp.simplify(v))


print("\nDistancia recorrida s(t):")


sp.pprint(sp.simplify(s))
```

### <span style="color: #1AA3D9"><center> Item 3: Evaluación numérica

Definiendo  a=0.9  evalúe  la  posición  de  la  partícula  durante  los  primeros 100s  de movimiento. En una gráfica <font color="#1AA3D9">tipo subplot</font> presente:

1. Trayectoria de la partícula (x vs y)
2. Velocidad en el tiempo (||v||vs t)
3. Distancia recorrida en el tiempo (distancia vs t)

```python
!pip install sympy
!pip install numpy 
!pip install matplotlib
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
import math

t = sp.Symbol("t, a")
a = 0.9
x = a**(t/math.pi) * sp.sin(t)
y = a**(t/math.pi) * sp.cos(t)
v = sp.sqrt(sp.diff(x, t)**2 + sp.diff(y, t)**2)

T = np.linspace(0, 100, 1000)

X, Y, V, D = [], [], [], []
dist = 0

for tt in T:
    vx = float(v.subs(t, tt))
    dist += vx 
    X.append(float(x.subs(t, tt)))
    Y.append(float(y.subs(t, tt)))
    V.append(vx)
    D.append(dist)

plt.figure(figsize=(10,10))
plt.subplot(3,1,1); plt.plot(X, Y); plt.title("Trayectoria"); plt.xlabel("x"); plt.ylabel("y")
plt.subplot(3,1,2); plt.plot(T, V); plt.title("Velocidad"); plt.xlabel("Tiempo"); plt.ylabel("Velocidad")
plt.subplot(3,1,3); plt.plot(T, D); plt.title("Distancia recorrida"); plt.xlabel("Tiempo"); plt.ylabel("Distancia")
plt.tight_layout(); plt.show()
```
